```python
import pandas as pd
import re
import numpy as np
```

```python
# extracted from https://www.cms.gov/Medicare/Coding/ICD9ProviderDiagnosticCodes/codes.html 
path2icd9 = '../data/ICD-9-CM-v32-master-descriptions/CMS32_DESC_LONG_SHORT_DX.xlsx'
path2subcodes = '../data/subcodes.csv'
icd9 = pd.read_excel(path2icd9)
```

```python
# build matching patterns for 3 digits category
def match_3digits(dxStr):
    if dxStr.startswith('E'):
        if len(dxStr) > 4: 
            return dxStr[4:] == '0'*len(dxStr[4:])
        else:
            # case 3digit category wo 0 ()
            #print('found Exxxx: {}'.format(dxStr))
            return True
    else:
        if len(dxStr) > 3: 
            return dxStr[3:] == '0'*len(dxStr[3:])
        else:
            #print('found xxx.xx: {}'.format(dxStr))
            return True

def build_3digits(dxStr):
    if dxStr.startswith('E'):
        if len(dxStr) > 4: 
            return dxStr[:4]
        else:
            return dxStr
    else:
        if len(dxStr) > 3: 
            return dxStr[:3]
        else:
            return dxStr
```

```python
subcodes = icd9[icd9.apply(lambda x: match_3digits(x['DIAGNOSIS CODE']), axis = 1)] 
```

```python
subcodes['3digit_code'] = icd9.apply(lambda x: build_3digits(x['DIAGNOSIS CODE']), axis = 1)
```

```python
# small test verifying that we did not parse two different codes in the same 3digit format
assert subcodes.shape[0] == len(np.unique(list(subcodes['3digit_code'].values)))
```
